```python
df_city = df.copy()
```

```python
# Вычислим общ.конверсию O2R
order_2_ride = df_full['ride_finish_time'].count()/df_full['order_id'].count()
print('Основная конверсия из заказа в поездку =',round(order_2_ride*100,2), 'процента')
```

```python
# Вычислим промежуточное звено конверсии O2A (как водители соглашаются на предложения)
order_2_assign = df_full['assign_time'].count()/df_full['order_id'].count()
print('Промежуточное звено конверсии,как водители соглашаются на предложения =',round(order_2_assign*100,2), 'процента')
```

```python
# Вычислим промежуточное звено конверсии A2A (как водители отменяют заказы после назначения)
assign_2_arrive = df_full['arrive_time'].count()/df_full['assign_time'].count()
print('Промежуточное звено конверсии,как водители отменяют заказы после назначения =',round(assign_2_arrive*100,2), 'процента')
```

```python
# Вычислим промежуточное звено конверсии A2R (как клиенты отменяют заказы после назначения)
arrive_2_ride = df_full['ride_finish_time'].count()/df_full['arrive_time'].count()
print('Промежуточное звено конверсии,как клиенты отменяют заказы после назначения =',round(arrive_2_ride*100,2), 'процента')
```

```python
# Создадим датафрейм и положим в него все вычисления
df_conv_order_avg = pd.DataFrame()
df_conv_order_avg = df_conv_order_avg.append({'city':'all',
                                              'order_2_ride':round(order_2_ride*100,2),
                                              'order_2_assign':round(order_2_assign*100,2),
                                              'assign_2_arrive':round(assign_2_arrive*100,2),
                                              'arrive_2_ride':round(arrive_2_ride*100,2)},ignore_index=True
                                            )                             
```

```python
df_conv_order_avg
```

```python
# Количество заказов в сутки по часам
number_of_hours = df_full.groupby(df_full['order_time'].dt.hour)['order_id'].count().reset_index()
number_of_hours 
```

```python
import matplotlib.pyplot as plt
```

```python
# Построим визуализацию
plt.figure(figsize=(10, 5)) # размерность
plt.title('Количество заказов') # название графика
plt.xlabel('Часы') # ось ОХ
plt.ylabel('Заказы') # ось ОУ
plt.plot(number_of_hours['order_time'],number_of_hours['order_id'] ) 
plt.grid() # сетка
plt.show()
```

```python
# среднее кол-во заказов в часы пик
df_conv_order_avg['order_avg_pic'] = int(number_of_hours.loc[number_of_hours['order_time'].isin((7,8,9,10,17,18,19,20))]['order_id'].mean())
```

```python
# среднее кол-во заказов в обычное время
df_conv_order_avg['order_avg'] = int(number_of_hours.loc[number_of_hours['order_time'].isin((0,1,2,3,4,5,6,11,12,13,14,15,16,21,22,23))]['order_id'].mean())
```

```python
# Разница в кол-ве заказов (пик.время и обыч.часы)
df_conv_order_avg['delta'] = df_conv_order_avg['order_avg_pic']- df_conv_order_avg['order_avg']
```

```python
# кол-во законченных зак-ов в сутки по часам
order_finish = df_full.groupby(df_full['ride_finish_time'].dt.hour)['order_id'].count().reset_index()
```

```python
# ср.кол-во законченных зак-ов в часы пик 
ord_fin_pic = int(order_finish.loc[order_finish['ride_finish_time'].isin((7,8,9,10,17,18,19,20))]['order_id'].mean())
```

```python
# ср.кол-во законченных зак-ов в обыч.время
ord_fin = int(order_finish.loc[order_finish['ride_finish_time'].isin((0,1,2,3,4,5,6,11,12,13,14,15,16,21,22,23))]['order_id'].mean())
```

```python
# конверсия законченных заказов в пик
df_conv_order_avg['o2r_fin_pic'] = round(ord_fin_pic/df_conv_order_avg['order_avg_pic']*100,2)
```

```python
# конверсия законченных заказов в обыч.время
df_conv_order_avg['o2r_fin'] = round(ord_fin/df_conv_order_avg['order_avg_pic']*100,2)
```

```python
# Посчитаем разницу в кол-ве заказов в часы пик и обычные часы по всем городам
# Посчитаем общ.конверсию по всем городам
# Вычислим разницу конверсий в часы пик и в обыч. час

x = (7,8,9,10,17,18,19,20)
y = (0,1,2,3,4,5,6,11,12,13,14,15,16,21,22,23)

result = []

for i in df_full.name_city.unique():
    
    df_loc = df_full[df_full['name_city']==i]# новый df с конкретным городом
    print(i)
    r1 = df_loc.groupby(df_loc['order_time'].dt.hour)['order_id'].count().reset_index()# кол-во заказов в каждый час
    r2 = int(r1.loc[r1['order_time'].isin((x))]['order_id'].mean())# ср.кол-во всех заказов в пик
    r3 = int(r1.loc[r1['order_time'].isin((y))]['order_id'].mean())# ср.кол-во всех заказов в обыч.время
    r4 = r2-r3
    
    r5 = df_loc['ride_finish_time'].count()/df_loc['order_id'].count()# общ.конверсия из заказа в поездку
    
    r6 = df_loc.groupby(df_loc['ride_finish_time'].dt.hour)['order_id'].count().reset_index()# кол-во законч.зак-ов в кажд.час
    r7 = int(r6.loc[r6['ride_finish_time'].isin((x))]['order_id'].mean())# ср.кол-во законченных заказов в пик
    r8 = int(r6.loc[r6['ride_finish_time'].isin((y))]['order_id'].mean())# ср.кол-во законченных заказов в обыч.время
    
    r9 = r7/r2 # общ.конверсия из заказа в поездку в пик
    r10 = r8/r3 # общ.конверсия из заказа в поездку в обыч.время    
    result.append({'city':i,
                   'delta_ord':r4,
                   'o2r':round(r5*100,2),
                   'o2r_fin_pic':round(r9*100,2),
                   'o2r_fin':round(r10*100,2),
                   'delta_o2r':round((r9-r10)*100,2)})
    print('Разница в количестве заказов в часы пик и в обычный час =',r4) 
    print('Основная конверсия из заказа в поездку =',round(r5*100,2), 'процента')
    print('Разница в конверсиях в часы пик и в обычный час =',round((r9-r10)*100,2), 'процента')
    print()
    
df_res = pd.DataFrame(result)    
df_res                 
    
```

```python
# Визуализация разницы кол-ва заказов по времени
plt.figure(figsize=(10, 5)) 

plt.title('Разница в кол-ве заказов между часами пик и обычным временем по городам')  
plt.xlabel('City') 
plt.ylabel('delta_ord') 
plt.xticks(rotation=45)

plt.bar(df_res['city'],df_res['delta_ord'])

plt.show() 
```

```python
# Визуализация общ.конверсии по городам
plt.figure(figsize=(15, 5))  
plt.title('Конверсия o2r по городам')  
plt.xlabel('City') 
plt.ylabel('o2r') 
plt.grid() 

plt.plot(df_res['city'], df_res['o2r'])

plt.show() 
```

```python
# Визуализация сравнение конверсий по времени в городах
plt.figure(figsize=(15, 5))  
plt.title('Конверсии в час пик и в обычное время')  
plt.xlabel('City') 
plt.ylabel('conv') 
plt.grid() 

plt.plot(df_res['city'], df_res['o2r_fin_pic'], color='r', label = 'O2R в пик')
plt.plot(df_res['city'], df_res['o2r_fin'], label = 'O2R в обычное время')

plt.legend()
plt.show() 
```

Изучите заказы в Хабаровске и Тюмени. Эти города интересуют менеджеров, так как в них наша компания начала работать относительно недавно.

 - На сколько процентных пунктов Order2Ride в среднем в них ниже, чем в других городах?
 
 - За счет какого звена воронки достигается эта разница? 
 
 - Сделайте выводы по Хабаровску и по Тюмени по отдельности.
 
 - Какие рекомендации вы можете дать локальным менеджерам?

```python
# Наш исходный датафрейм
df_full.head()
```

```python
# Посчитаем общую конверсию и ее звенья по всем городам
conv = []

for i in df_full.name_city.unique():
    
    df_con = df_full[df_full['name_city']==i]# новый df с конкретным городом
    
        
    a1 = df_con['ride_finish_time'].count()/df_con['order_id'].count()# общ.конверсия из заказа в поездку
    a2 = df_con['assign_time'].count()/df_con['order_id'].count()# конв. order_2_assign, как водители соглашаются на предложения
    a3 = df_con['arrive_time'].count()/df_con['assign_time'].count()# конв.assign_2_arrive,как водители отменяют заказы после назначения
    a4 = df_con['ride_finish_time'].count()/df_con['arrive_time'].count()# конв. arrive_2_ride, как клиенты отменяют заказы после назначения
    
    conv.append({'city':i,
                 'o2r':round(a1*100,2),
                 'order_2_assign':round(a2*100,2),
                 'assign_2_arrive':round(a3*100,2),
                 'arrive_2_ride':round(a4*100,2)})

df_con_city = pd.DataFrame(conv)    
df_con_city              
 
```

```python
# Вычислим разницу O2R в Тюмени в сравнении с другими городами, положим данные в df_res
tum = []

for i in df_con_city['city']:
    t = int(df_con_city[df_con_city['city']==i]['o2r'])-int(df_con_city.loc[df_con_city['city']=='Тюмень']['o2r'])
    tum.append(t)
df_con_city['tum_del'] = tum   
```

```python
# Вычислим разницу O2R в Хабаровске в сравнении с другими городами, положим данные в df_res
hab = []

for i in df_con_city['city']:
    h = int(df_con_city[df_con_city['city']==i]['o2r'])-int(df_con_city.loc[df_con_city['city']=='Хабаровск']['o2r'])
    hab.append(h)
df_con_city['hab_del'] = hab   
```

Анализ воронки **Order2Ride** во всех городах показывает, что конверсия в поездку в Тюмени и Хабаровске значительно отличается по причине 
низкой конверсии в промежуточных звеньях воронки.

В Тюмени низкая конверсия в звене **order_2_assign** (из заказа в назначение водителя) и низкая конверсия в звене **arrive_2_ride** (из-за отмен заказов клиентами, уже после прибытия водителя к пассажиру).

В Хабаровске мало отмен заказов клиентами, но самая низкая конверсия в назначение водителя. Возможно это связано с недостаточным количеством автомобилей на линии.

```python
import scipy.stats as st
```

**Вычислим доверительные интервалы** 

```python
# Вычислим доверительные интервалы для звеньев конверсии O2R тарифа "Доставка" 
l_o2a, r_o2a = st.t.interval(alpha=0.95, df=len(df_con_city_dos['order_2_assign']-1),
              loc=np.mean(df_con_city_dos['order_2_assign']),
              scale=st.sem(df_con_city_dos['order_2_assign']))
print('Доверительный интервал assign_2_arrive "Доставка" =',l_o2a,'-',r_o2a)
df_con_city_dos['l_o2a'] = l_o2a
df_con_city_dos['r_o2a'] = r_o2a
```

```python
l_a2a, r_a2a = st.t.interval(alpha=0.95, df=len(df_con_city_dos['assign_2_arrive'])-1,
              loc=np.mean(df_con_city_dos['assign_2_arrive']),
              scale=st.sem(df_con_city_dos['assign_2_arrive']))
print('Доверительный интервал assign_2_arrive "Доставка" =',l_a2a,'-',r_a2a)
df_con_city_dos['l_a2a'] = l_a2a
df_con_city_dos['r_a2a'] = r_a2a
```

```python
l_a2r, r_a2r = st.t.interval(alpha=0.95, df=len(df_con_city_dos['arrive_2_ride'])-1,
              loc=np.mean(df_con_city_dos['arrive_2_ride']),
              scale=st.sem(df_con_city_dos['arrive_2_ride']))
print('Доверительный интервал arrive_2_ride "Доставка" =',l_a2r,'-',r_a2r)
df_con_city_dos['l_a2r'] = l_a2r
df_con_city_dos['r_a2r'] = r_a2r
```

```python
# Посчитаем общую конверсию для всех тарифов по всем городам 
O2R = []

for i in df_full.name_city.unique():
    df_tariff = df_full[(df_full['name_city']==i) & (df_full['name_tariff']=='Бизнес')]# новый df с конкретным городом и тарифом      
    t1 = df_tariff['ride_finish_time'].count()/df_tariff['order_id'].count()# общ.конверсия из заказа в поездку
        
    df_tariff = df_full[(df_full['name_city']==i) & (df_full['name_tariff']=='Эконом')]
    t2 = df_tariff['ride_finish_time'].count()/df_tariff['order_id'].count()# общ.конверсия из заказа в поездку
    
    df_tariff = df_full[(df_full['name_city']==i) & (df_full['name_tariff']=='Комфорт')]
    t3 = df_tariff['ride_finish_time'].count()/df_tariff['order_id'].count()# общ.конверсия из заказа в поездку
    
    df_tariff = df_full[(df_full['name_city']==i) & (df_full['name_tariff']=='Комфорт+')]
    t4 = df_tariff['ride_finish_time'].count()/df_tariff['order_id'].count()# общ.конверсия из заказа в поездку
    
    df_tariff = df_full[(df_full['name_city']==i) & (df_full['name_tariff']=='Доставка')]
    t5 = df_tariff['ride_finish_time'].count()/df_tariff['order_id'].count()# общ.конверсия из заказа в поездку
    
    
            
    O2R.append({'city':i,
                'o2r_бизнес':round(t1*100,2),
                'o2r_экон':round(t2*100,2),
                'o2r_комф':round(t3*100,2),
                'o2r_комф_плюс':round(t4*100,2),
                'o2r_дост':round(t5*100,2)})
         
  
df_tariff_conv = pd.DataFrame(O2R)    
df_tariff_conv           
```
